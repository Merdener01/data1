```python

def mean_absolute_percentage_error(y, ypred):
  return np.mean(np.abs(y - ypred) / np.abs(y)) * 100
```

```python
ypred = tempmax.rolling(2).mean().shift(1)

plt.plot(tempmax, "-ob", label = "observed")
plt.plot(ypred, "-*r", label = "forecasted")
plt.grid()
plt.legend()

print(mean_absolute_percentage_error(tempmax, ypred))

mask = ypred.notna() & (tempmax != 0)   # NaN ve s覺f覺rlar覺 d覺la
mape = mean_absolute_percentage_error(tempmax[mask], ypred[mask])
print("MAPE:", mape, "(", mape*100, "% )")
```

```python
trend =[None, "add", "mul"]
damp =[False, True]
seasonal =[None, "add", "mul"]

best_mape = np.inf
best_params = (None, False, None) # Initialize best_params

for t in trend:
  for d in damp:
    for s in seasonal:
      if t is None and d is True: # Damped trend is only applicable when trend is not None
        continue
      try:
        mdl = ExponentialSmoothing(tempmax_train, trend=t, damped_trend=d, seasonal=s,
                                  seasonal_periods=12).fit()

        ypred = mdl.predict(tempmax_test.index[0],tempmax_test.index[-1])
        mape = mean_absolute_percentage_error(tempmax_test, ypred)

        if mape < best_mape:
          best_mape = mape
          best_params = (t, d, s)
      except:
        # Handle potential errors during model fitting for certain parameter combinations
        continue

print(best_params)
```
